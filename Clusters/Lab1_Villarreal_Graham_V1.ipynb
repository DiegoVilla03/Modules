```python
import numpy as np
import clusters as cl
```

## Example of Usage

the definiton of data

```python
means = {
    '1': np.array([0, 0, 0]),
    '2': np.array([3, 3, 3]),
    '3': np.array([0, 3, 3])
}

covariances = {
    '1': np.array([[1, 0.5, 0.3], [0.5, 1, 0.2], [0.3, 0.2, 1]]),
    '2': np.array([[1, 0.3, 0.1], [0.3, 1, 0.2], [0.1, 0.2, 1]]),
    '3': np.array([[1, 0.4, 0.2], [0.4, 1, 0.3], [0.2, 0.3, 1]])
}

samples = {
    '1': 100,
    '2': 150,
    '3': 200
}
```

```python
# object inicialization 
sim1 = cl.generate_clusters.from_multivariate_normal(means, covariances, samples)

# 2d ploting of data
# replace parameter lables to False for data only 
sim1.plot_2d(labels=True)
```

```python
# 3d ploting of data
# replace parameter lables to False for data only 
sim1.plot_3d(labels=True)
```

```python
# convert the data into a df to operate with pandas 
# Replace parameter labels to False for data only 
sample_data = sim1.to_df(labels=True)

sample_data.head()
```

```python
sim1.pairplot(labels=True, palette='summer')
```

```python
sim1.get_points
```

```python
sim1.get_dimensions
```

### Example of usage 

Uniform samples

```python
ranges = {
    0: (np.array([0, 0, 0]), np.array([1, 1, 1])),  # Rango de (0,0,0) a (1,1,1) para el grupo 0
    1: (np.array([2, 2, 2]), np.array([4, 4, 4])),  # Rango de (2,2,2) a (4,4,4) para el grupo 1
    2: (np.array([-1, -1, -1]), np.array([0, 0, 0])) # Rango de (-1,-1,-1) a (0,0,0) para el grupo 2
}

# Definir tamaños de muestra para cada grupo
sample_sizes = {
    0: 100,  # 100 muestras para el grupo 0
    1: 150,  # 150 muestras para el grupo 1
    2: 50    # 50 muestras para el grupo 2
}
```

```python
sim2 = cl.generate_clusters.from_uniform(ranges, sample_sizes)
```

```python
sim2.plot_2d(labels=True)
```

```python
sim2.plot_3d(labels=True)
```

```python
sample_data = sim2.to_df(labels=True)

sample_data.head()
```

```python
sim2.pairplot(labels=True, palette='plasma')
```
