```python
import clusters as cl
import numpy as np
```

Se generan datos aleatorios para inicializar la clase

```python
means = {
    '1': np.array([0, 0, 0]),
    '2': np.array([3, 3, 3]),
    '3': np.array([0, 3, 3])
}

covariances = {
    '1': np.array([[1, 0.5, 0.3], [0.5, 1, 0.2], [0.3, 0.2, 1]]),
    '2': np.array([[1, 0.3, 0.1], [0.3, 1, 0.2], [0.1, 0.2, 1]]),
    '3': np.array([[1, 0.4, 0.2], [0.4, 1, 0.3], [0.2, 0.3, 1]])
}

samples = {
    '1': 100,
    '2': 150,
    '3': 200
}
```

```python
# constructor inicial para generar datos con distribución Normal Multivariada

sim1 = cl.generate_clusters.from_multivariate_normal(means, covariances, samples)
```

```python
data = sim1.to_df(labels=False)
data.head()
```

```python
# se selecciona un punto arbitrario para iniciar el calculo de metricas
point = np.array([1,1,1])

# Distancia de un punto a todos los puntos en el conjunto
# la funcion recibe un punto en la dimension de los datos y una metrica, la default es 'Euclidean'
### Ejemplo de metrica es 'cosine', 'Manhattan'
dist_point_to_all = sim1.distance_point_to_all(point)

print(dist_point_to_all)
```

```python
# Distancia de un punto a un subconjunto de puntos
subset_indices = np.random.choice(sim1.dist_no_labels.shape[0], 10, replace=False)
# La funcion recibe un punto dado por el usuario en la dimensión de los datos, un array de indices que conforman un subconjunto de los datos y una medida de proximidad de el punto con el subconjunto 
dist_point_to_subset = sim1.distance_point_to_subset(point, subset_indices, measure='min')

print(dist_point_to_subset)
```

```python
# Distancia entre dos subconjuntos de puntos
subset1_indices = np.random.choice(sim1.dist_no_labels.shape[0], 10, replace=False)
subset2_indices = np.random.choice(sim1.dist_no_labels.shape[0], 15, replace=False)
dist_subset_to_subset = sim1.distance_between_subsets(subset1_indices, subset2_indices, measure='min')

print(dist_subset_to_subset)
```
